```python

def gen_reading_frames(seq):
    frames = []
    # gives three reading frames
    frames.append(translate_seq(seq, 0))
    frames.append(translate_seq(seq, 1))
    frames.append(translate_seq(seq, 2))
    frames.append(translate_seq(reverse_complement(seq), 0))
    frames.append(translate_seq(reverse_complement(seq), 1))
    frames.append(translate_seq(reverse_complement(seq), 2))
    return frames
```

```python

def Fibonacci_Loop(number):
    old = 1
    new = 1
    for itr in range(number - 1):
        tmpval = new
        new = old
        old = old + tmpval
    return new

# tmpval        1   1   2   3
# new       1   1   2   3   4
# old       1   2   3   4   7

def Fibonacci_Loop_Pythonic(number):
    old, new = 1, 1
    for itr in range(number - 1):
        new, old = old, old + new
    return new

# new   1   1   2   3   5 
# old   1   2   3   5   8


Fibonacci_Loop_Pythonic(20)
```

```python
# Rabbits and Recurrence Relations

def Fibonacci_Loop_Rabbits(months, offsprings):
    parent, child = 1, 1
    for itr in range(months - 1):
        child, parent = parent, parent + (child * offsprings)
    return child

Fibonacci_Loop_Rabbits(35, 2)
```

```python
# Given: Three positive integers k, m, and n, representing a population containing k+m+n
# organisms: k individuals are homozygous dominant for a factor, 
# m are heterozygous, 
# and n are homozygous recessive.

# Return: The probability that two randomly selected mating organisms will produce an individual possessing 
# a dominant allele (and thus displaying the dominant phenotype). Assume that any two organisms can mate.

#     het + het
#   
#     |  A | a
#   -----------
#   A | AA | Aa
#   a | Aa | aa


#     het + rec
#   
#     |  a | a
#   -----------
#   A | Aa | Aa
#   a | aa | aa

# all combinations of K+K and K+M would have at least one dominant allele
# 0.75 combinations of M+M would have at least one dominant allele
# 0.5 combinations of M+N would have at least one dominant allele
# 0 combinations of N+N have a dominant allele

from scipy.special import comb

k = 17
m = 25
n = 26

# total number of organisms in the population
tot_pop = k + m + n

# calculate the number of combos that can be made
tot_comb = comb(tot_pop, 2)

# calculate the number of combos that have a dominant allele
val_comb = comb(k,2) + k*m + k*n + 0.5*m*n + 0.75*comb(m,2)

prob = val_comb / tot_comb

print(f'The probability that an individual will have at least one dominant allele:\n{prob}')
```

```python
# Mortal Fibonacci Rabbits
# Given: Positive integers n≤100 and m≤20.
# Return: The total number of pairs of rabbits that will remain after the n-th month if all rabbits live for m months.

n = 99    # how many months we are counting for    #6
m = 17    # how long each pair of rabbit lives     #3
# answer = 4

def mortal_rabbits (n: int, m: int) -> int:
    if n == 0:
        return 0
    if n == 1 or n ==2:
        return 1
# we track the number of rabbits of each age (from 1 to m months old)
# ages[i-1] = number of rabbits that are i months old at the end of the current month

    ages = [0] * m     # creates a list of m elements that are all zeroes
    ages[0] = 1    # month 1: one newborn pair (age 1)

    for month in range(2, n+1):
        new_ages = [0] * m
    
        # all rabbits that were alive age by 1 month
        for age in range(1, m):
            new_ages[age] = ages[age-1]

# newborn rabbits (age 0 or element 0 in list) 
# are produced by mature pairs (those at least 2 months old or ages[1:])
        new_ages[0] = sum(ages[1:])  #if we exclude dying rabbits the sum(ages[1:m-1])

# the oldest rabbits (age m) die and are not carried over because we exclude ages[m-1]
        ages = new_ages

    return sum(ages)

result =mortal_rabbits(n,m)
print(result)
```

```python
# Overlap Graphs
# Given: A collection of DNA strings in FASTA format having total length at most 10 kbp.
# Return: The adjacency list corresponding to O3. You may return edges in any order.

from Bio import SeqIO
# current_dir = Path.cwd()
data_dir = 'C:/Users/ianj_/OneDrive/Documents/coding_projects/rosalind/fasta_file.txt'

# parse the fasta file and map each record ID to its DNA sequence
# SeqIO.parse returns an iterator for seq record objects, we convert it to a dictionary for easier access
records = SeqIO.to_dict(SeqIO.parse(data_dir, 'fasta'))

# extract the list of labels (record ID's) from the records
# these will be the node names in the graphs

labels = list(records.keys())

# initialize an empty list to store the edges of the adjacency list
# each edge will be a tuple (from_edge, to_edge)
edges = []
k = 3

# iterate over every possible pair of sequences (i != j) to check for overlaps
for i in range(len(labels)):
    from_label = labels[i]
    from_seq = str(records[from_label].seq)   # convert seq object to string to allow slicing

    for j in range(len(labels)):
        if i == j:
            continue   # skip self loops
        to_label = labels[j]
        to_seq = str(records[to_label].seq)

        # check if the last k characters of from_seq exactly match the first k characters of to_seq
        # if yes there is a directed edge from freom_lable to to_label
        if from_seq[-k:] == to_seq[:k]:
            edges.append((from_label, to_label))

# returns a list of edges, each edge represents a directed connection in the overlap graph 03 (k=3)
# converts list of tuples into two columns without commas or brackets
for from_label, to_label in edges:
    print(from_label, to_label)
```

```python
# Calculating Expected Offspring

# Given: Six nonnegative integers, each of which does not exceed 20,000. 
# The integers correspond to the number of couples in a population possessing each genotype pairing for a given factor. 
# In order, the six given integers represent the number of couples having the following genotypes:
# AA-AA
# AA-Aa
# AA-aa
# Aa-Aa
# Aa-aa
# aa-aa
# Return: The expected number of offspring displaying the dominant phenotype in the next generation, 
# under the assumption that every couple has exactly two offspring.

#   Common Probability Ratios (Mendelian Genetics)
#   AA x AA: 100% AA (100% Dominant)
#   AA x Aa: 100% Dominant (50% AA, 50% Aa)
#   AA x aa: 100% Dominant (Aa)
#   Aa x Aa: 75% Dominant (25% AA, 50% Aa), 25% Recessive (aa)
#   Aa x aa: 50% Dominant (Aa), 50% Recessive (aa)
#   aa x aa: 100% Recessive (aa)


raw_data = '17374 19277 18429 16311 18505 16801'
data = raw_data.strip().split()    # convert raw data into string element list
couples = [int(x) for x in data]   # convert data into integers

# unpack the list into meaningful variables
AA_AA, AA_Aa, AA_aa, Aa_Aa, Aa_aa, aa_aa = couples

# initialize the total dominant
total_dominant = 0

# each couple produces two offspring
# calculate the number of dominant phenotype offspring

# 100% of AA_AA are dominant
total_dominant += AA_AA * 2 * 1     # 100% are dominant
total_dominant += AA_Aa * 2 * 1     # 100% are dominant
total_dominant += AA_aa * 2 * 1     # 100% are dominant
total_dominant += Aa_Aa * 2 * 0.75  # 75% are dominant
total_dominant += Aa_aa * 2 * 0.5   # 50% are dominant
total_dominant += aa_aa * 2 * 0     # 0% are dominant

# output the final result
print(total_dominant)
```

```python
# Finding a Shared Motif
# Given: A collection of k (k≤100) DNA strings of length at most 1 kbp each in FASTA format.
# Return: A longest common substring of the collection. (If multiple solutions exist, you may return any single solution.)


# read the raw data file
data_dir = 'C:/Users/ianj_/OneDrive/Documents/coding_projects/rosalind/overlap_data.txt'
with open(data_dir, 'r') as file:
    data = file.read()

def read_fasta(data):
    """Parse fasta file and return a list of DNA strings (without headers)"""
    dna_strings = []
    current = ''
    for line in data.strip().splitlines():
        if line.startswith('>'):
            if current:
                dna_strings.append(current)
            current = ''
        else:
            current += line.strip()
    if current:
        dna_strings.append(current)
    return dna_strings


def longest_common_substring(dna_strings):
    """find and return one longest common substring from a list of strings"""
    #dna_strings = read_fasta(data)
    if not dna_strings:
        return ''
    # the longest common substring can't be longer than the shortest string
    shortest = min(dna_strings, key=len)
    n = len(shortest)
    # try all substrings of shortest string
    for length in range(n, 0, -1):    # start with the longest possible length
        for i in range(n - length + 1):
            candidate = shortest[i : i + length]    # at each starting position i extract a substring of current length, all substrings of current length from the shortest string

            # check if the candidate is present in other strings
            if all(candidate in dna for dna in dna_strings):
                return candidate    # found the shortest common substring

    return ''

dna_strings = read_fasta(data)
result = longest_common_substring(dna_strings)

print(result)
```

```python
# Independent Alleles

# Given: Two positive integers k (k ≤ 7) and N (N ≤ 2k). 
# In this problem, we begin with Tom, who in the 0th generation has genotype Aa Bb. Tom has two children in the 1st generation, 
# each of whom has two children, and so on. Each organism always mates with an organism having genotype Aa Bb.

# Return: The probability that at least N Aa Bb organisms will belong to the k-th generation of Tom's family tree 
# (don't count the Aa Bb mates at each level). Assume that Mendel's second law holds for the factors.

from math import comb

# enter the input, two numbers k and N
# 2 1    are the values in the example
k, N = map(int, input().split())

# number of organisms in k-th generation, each has two children so 2^k
m = 2 ** k

# probability p that any organism in gen k is Aa Bb
p = 0.25    # since 1/2 * 1/2 = 1/4

# we need probability P(X >= N) where X ~ binomial(M, P)
# compute using sum of binomial probabilities
prob = 0
# loops over all possible organisms from N to m (inclusive)
# this calculates the right tail probability of P(X >= N)
for x in range(N, m + 1):
    # C(m, x) * p^x * (1-p)^(m-x)
    # comb(m, x) → Number of ways to choose which x organisms out of m are Aa Bb (the binomial coefficient "m choose x").
    # p ** x → Probability that those x specific organisms are Aa Bb (each has probability 0.25).
    # (1 - p) ** (m - x) → Probability that the remaining (m - x) organisms are not Aa Bb (each has probability 0.75).
    term = comb(m, x) * (p ** x) * ((1 - p) ** (m - x))
    prob += term

# probability is the output
print(f"{prob:.3f}")
```

```python
# Open Reading Frames
# Given: A DNA string s of length at most 1 kbp in FASTA format.
# Return: Every distinct candidate protein string that can be translated from ORFs of s. 
#   Strings can be returned in any order.

# create a dictionary to define the genetic code (nucleatides -> proteins)
# stop codons are None or *
codon_table = {
    'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M',
    'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T',
    'AAC':'N', 'AAT':'N', 'AAA':'K', 'AAG':'K',
    'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R',
    'CTA':'L', 'CTC':'L', 'CTG':'L', 'CTT':'L',
    'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P',
    'CAC':'H', 'CAT':'H', 'CAA':'Q', 'CAG':'Q',
    'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R',
    'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
    'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A',
    'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E',
    'GGA':'G', 'GGC':'G', 'GGG':'G', 'GGT':'G',
    'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S',
    'TTC':'F', 'TTT':'F', 'TTA':'L', 'TTG':'L',
    'TAC':'Y', 'TAT':'Y', 'TAA':'*', 'TAG':'*',
    'TGC':'C', 'TGT':'C', 'TGA':'*', 'TGG':'W',
}

def reverse_complement(dna):
    """Return reverse complement of a DNA string."""
    complement = {'A':'T', 'T':'A', 'C':'G', 'G':'C'}
    #Reverse the string and replace with its complement
    return ''.join(complement[base] for base in reversed(dna))

def translate(dna):
    """Translate a DNA sequence into a protein sequence until a stop codon is reached."""
    protein = []
    for i in range(0, len(dna) - 2, 3):   #step by three nucleotides (codons)
        codon = dna[i : i + 3]
        if len(codon) < 3:
            break
        aa = codon_table.get(codon, 'X')   #X is for unknown codon which should not happen
        if aa == '*':
            #break   #end translation at stop codon
            return ''.join(protein), True   #Stop is found
        protein.append(aa)
    return ''.join(protein), False   #No stop is found 
            
def find_orfs(dna):
    """Find all possible protein strings from ORF's in a DNA string."""
    proteins = set()   #Use set to automatically handle duplicates
    n = len(dna)

    #Check all 3 reading frames
    for frame in range(3):
        #Scan for start codons (ATG) in this frame
        for i in range(frame, n - 2, 3):
            if dna[i : i + 3] == 'ATG':
                #translate from this codon onwards
                protein, has_stop = translate(dna[i :])
                if protein and has_stop:   #only add if there is a protein and a stop codon
                    proteins.add(protein)
    return proteins

#Read the input file and extract a DNA sequence
data_dir = 'C:/Users/ianj_/OneDrive/Documents/coding_projects/rosalind/orf_data.txt'
with open(data_dir, 'r') as file:
    input_data = [line.strip() for line in file.readlines() if line.strip()]
    dna = ''.join(line.strip() for line in input_data[1:] if line.strip())

#Get proteins from the forward DNA strain
proteins = find_orfs(dna)

#Get proteins from the reverse complement strand
rev_dna = reverse_complement(dna)
proteins.update(find_orfs(rev_dna))

#Write the output, one protein per line
for prot in sorted(proteins):
    print(prot)
```

```python
# A Simple Measure of Gene Order Similarity
# Given: A positive integer n≤10000 followed by a permutation π of length n
# Return: A longest increasing subsequence of π, followed by a longest decreasing subsequence of π.

import sys

def find_lis(arr):
    n = len(arr)
    if n == 0:
        return []
    # tails stors the smailest tail of subsequences i+1
    tails = []
    # prev[i] = previous index of the LIS ending at i
    prev = [-1] * n
    # tail_indices to reconstruct — actually we use indices in tails list
    for i in range(n):
        # Binary search to find the position where arr[i] should be placed in tails
        left, right = 0, len(tails) - 1
        while left <= right:
            mid = (left + right) // 2
            if arr[tails[mid]] < arr[i]:    # strict increasing
                left = mid + 1
            else:
                right = mid - 1
        pos = left

        # extend or replace the tail
        if pos == len(tails):
            tails.append(i)
        else:
            tails[pos] = i

        # record the predecessor for reconstruction
        if pos > 0:
            prev[i] = tails[pos - 1]
        else:
            prev[i] = -1
    # Reconstruct the LIS by backtracking from the end
    lis = []
    if tails:
        last = tails[-1]
        while last != -1:
            lis.append(arr[last])
            last = prev[last]
        lis.reverse()
    return lis

#n = find_lis('51423')   #must feed in parameter as a string because len() does not work with int type
#print(n)

# tails maintains the smallest possible ending value (via its index) for all increasing subsequences of each length.
# We use binary search (left/right) to find where the current element arr[i] fits in O(log n) time.
# prev array links each position to the previous one in its current best subsequence.
# At the end, we follow the prev pointers from the longest tail to build the actual subsequence and reverse it.




def find_lds(arr):
    n = len(arr)
    if n == 0:
        return []
    tails = []
    prev = [-1] * n

    for i in range(n):
        left, right = 0, len(tails) -1
        while left <= right:
            mid = (left + right) // 2
            if arr[tails[mid]] > arr[i]:   #strict decreasing
                left = mid + 1
            else:
                right = mid - 1
        pos = left

        if pos == len(tails):
            tails.append(i)
        else:
            tails[pos] = i

        if pos > 0:
            prev[i] = tails[pos - 1]
        else:
            prev[i] = -1

    lds = []
    if tails:
        last = tails[-1]
        while last != -1:
            lds.append(arr[last])
            last = prev[last]
        lds.reverse()
    return lds

#n = find_lds('51423')   #must feed in parameter as a string because len() does not work with int type
#print(n)

# Almost identical to find_lis, except the comparison is reversed (> instead of <) so it finds the longest strictly decreasing subsequence.

# all input data should be on one-line string
input = '8137 1049 3834 7494 967 3977 442 2369 7831 592 6809 4497 3120 6071 5445 4596 3493 5429 1052 2824 2134 1092 754 5107 708 7474 5350 3736 5301 7165 3825 5361 3855 2842 4023 5696 5444 5880 2879 6331 6380 7384 6129 2943 2841 8012 7477 171 6073 4328 6860 3355 2185 6794 6895 7745 1644 529 3781 3851 3589 1778 6159 3853 4958 1430 4836 1526 234 1243 1077 3931 6358 7056 4605 5133 7434 7526 7543 3874 6669 7021 6948 2849 1598 296 5417 3865 1612 3381 4648 7682 938 4773 1727 6732 4536 3354 7766 6246 1569 2844 1673 8004 4468 4450 4427 2419 5119 4312 2893 2997 7816 3696 2139 5066 742 1281 739 4751 1019 669 5891 3974 2128 7681 7033 776 6988 3961 1191 3809 913 2679 7183 1496 1492 3452 2426 2344 7968 7306 3976 2838 6532 7913 2509 4072 6542 4136 5701 3480 6935 6284 2036 6204 1783 3129 7469 2279 3786 1262 558 2044 5936 7425 7503 1469 1380 4509 3439 31 3424 1965 5489 906 4028 5868 541 1736 2319 3532 3991 8059 6597 7659 957 1489 1794 7454 4032 2313 7403 4358 4290 4285 490 3998 7308 4781 3595 5547 4407 1431 4057 5691 1542 5982 2058 3700 3501 2029 6631 52 1214 756 2052 3821 7117 3594 4317 7072 7568 5527 4822 5216 4244 4652 5411 6317 4580 5981 4441 7078 5581 4563 3705 5774 5643 3310 2000 8022 617 5751 5338 5720 3707 1982 1351 1525 4641 6641 7211 6965 1335 4817 5934 5495 6379 1278 7310 7372 1660 5336 6151 2555 7691 7468 7825 3258 2839 3752 2151 1181 4890 7146 3535 1841 6553 4399 5164 6913 5498 2182 591 5574 4722 4884 6448 333 3971 1046 6282 3854 5440 7858 6109 4179 6658 1158 1247 2158 8128 5001 6914 4415 5211 1547 6568 4800 1197 3210 3122 7184 3206 275 1836 2008 4364 5702 4830 661 3262 5519 2702 5550 91 7948 2796 2370 1344 3108 2773 2717 5217 5170 5593 1456 608 5590 4288 7660 1024 1491 2422 457 2322 4695 2181 6405 6555 137 3311 4344 3624 1800 943 6228 4316 7237 7542 6138 4718 5821 1769 1786 1734 607 660 5219 5319 5187 1356 1078 6825 4004 3812 6857 5714 5632 3544 5115 862 7198 5532 5451 849 6890 4395 7267 5330 5389 2490 2267 213 4154 3161 320 7741 7453 3512 5949 2064 2353 4838 3253 770 646 1575 3975 2595 3878 3132 2575 2377 2361 2366 5896 4439 6511 1597 2356 6299 4105 3170 44 1726 3575 2953 3292 1930 1720 7712 3619 10 4070 3724 5110 979 4698 6060 3799 4390 7801 7684 5218 5181 2315 7517 7320 2614 7330 2865 2999 931 2502 3620 3288 5582 443 3505 2629 2165 6636 447 1124 713 6581 1018 3753 3859 3348 7062 2721 5393 2135 7960 7786 5446 3431 674 1564 5813 5055 200 4573 6633 5732 6868 3165 5911 1655 3303 1966 2640 5464 6765 5226 149 5436 7231 6670 4828 635 1942 7943 3254 4246 3727 3379 6290 4159 150 3455 4082 6601 988 4874 1875 4058 7719 6759 4533 1293 4209 7161 2244 7092 163 5140 134 853 856 4143 4147 876 5927 2925 7915 8011 7512 430 3091 3119 7985 3939 6875 4408 4249 7770 1336 1141 2700 5273 1099 7369 7128 3582 6446 5698 332 5255 614 7153 6187 637 6293 7833 8117 7401 7224 7204 3793 4600 226 4024 6979 2435 993 5798 5512 5989 2857 8131 4952 5997 7553 732 2334 3884 4770 4920 3774 6667 5860 218 7068 5034 343 7298 3744 7844 7571 6569 54 5910 1908 5864 3740 3605 28 994 1893 573 7859 7332 4191 5708 2998 7534 2866 1385 4714 2078 3327 5493 6891 2788 4411 2818 659 7897 4491 4515 4589 1831 4434 6678 2083 4340 777 5585 2184 6889 6320 7028 3445 590 6340 1999 4134 4807 153 2726 6543 3291 7679 6198 6031 842 4894 6680 4963 3922 366 3963 1960 5328 5021 5155 5570 1739 2916 1333 4039 3391 7452 681 336 6493 947 1455 71 4592 4231 6519 3890 1693 3477 5583 7735 4391 1641 5296 4710 2367 4189 2320 7060 1264 8125 4623 177 230 1441 583 6408 1383 2690 5734 1593 3315 6839 4686 3053 7713 2172 5564 6309 3677 5102 7608 6103 684 3085 6002 5232 7814 6587 946 6877 345 2505 6805 7665 2670 7074 6565 1323 4819 2225 6250 5265 810 1486 4239 3682 76 3025 2918 6489 2062 37 3691 670 6304 1584 1533 7422 5879 3524 7163 3811 1000 4564 7975 7730 4579 5041 133 3668 7007 2596 1180 3322 1268 5571 6423 3946 3540 1034 5999 8020 5885 4264 7656 7191 7923 4176 1847 1317 3231 531 1445 2014 2429 3465 1116 6065 2763 5432 6210 3107 3615 1144 722 7109 43 865 889 3075 5636 5617 1695 410 4859 1051 4578 5747 507 6188 6961 6854 2529 5929 351 3921 6942 6040 2465 2227 3335 4728 7050 5443 5462 6826 4252 101 7611 1295 3278 923 331 5757 6674 110 2251 289 7484 3383 8009 3474 1398 4701 4081 6884 4861 535 6563 7998 3333 4606 6402 7794 7685 5309 6175 3580 4462 2514 7605 2728 3421 8051 4767 3943 3167 2148 2604 5584 4691 7055 2636 2198 5744 7154 839 1215 411 5450 2123 1912 3866 1929 4977 6650 840 6116 4091 1520 7817 6740 1777 7136 795 465 1235 549 5912 7354 5656 743 6644 3329 7236 6352 7044 6583 1166 6006 4904 8088 6876 6437 5673 4073 3766 2531 182 4661 6567 513 7803 1756 1022 5820 769 462 5869 2933 6733 2411 1417 6962 3010 7795 2482 2345 3248 4849 2736 7229 1156 1896 7010 4929 1997 3651 4303 6726 4145 3499 7032 3838 2897 2864 2156 6621 324 2519 4997 7087 67 4465 6584 4843 524 3622 1895 6249 815 3344 445 5358 6227 3169 5439 7001 6417 1058 2710 7775 5730 7160 1484 6098 701 1159 5225 5537 4377 7578 4308 3003 2969 2027 1975 2671 5261 552 5239 7197 1621 6710 5601 6753 1465 2747 4516 6349 3287 5254 5307 574 5748 2518 7446 4302 7958 4051 1955 7582 4946 2693 1447 2300 2343 5985 1112 6322 7254 7127 7115 7335 7966 379 6343 6672 48 4141 2928 4382 2124 2152 7289 8101 1515 467 7560 7604 2961 3527 2993 2180 2417 2892 6855 2587 1412 3282 8034 3155 7029 7245 3986 7546 2520 5471 1860 5665 2485 1482 5042 7397 5587 3033 7220 1939 1244 2696 6608 508 7363 7475 6527 1816 4524 6830 6516 7762 7938 4614 6521 6474 2942 766 6288 7421 4410 4750 764 7118 881 3044 420 6119 6348 2188 1632 3502 835 3515 3006 5421 7284 6295 172 4692 3469 4895 4085 3293 7031 1882 1410 2977 2254 5920 5678 1815 4812 6362 4989 3398 5195 2019 3640 7941 7040 5442 5420 6426 3800 3933 2955 7121 2856 4662 6870 6170 3617 1357 1943 4149 5416 3076 6739 1205 6957 4996 986 4687 654 1033 2873 6795 3102 1487 4960 7449 87 7271 4525 860 3353 4585 783 2821 7034 1100 3486 6960 2380 1913 2905 3130 3616 6993 6297 7490 7842 1009 7030 7417 5018 5413 897 8071 3414 5830 2526 7673 7891 1651 2226 2107 1782 4546 1961 70 6562 2252 5791 6917 32 580 1827 2222 1604 2452 2532 2698 977 7972 2238 5888 2858 7111 6529 4544 640 7466 4607 5773 404 2331 6679 1507 5972 6497 5877 980 7839 2290 1894 7661 2574 3034 3260 1305 5023 1360 7246 3400 3760 2935 3858 8065 3792 4850 6000 6506 2430 2753 2913 3670 8018 3969 5166 4094 618 6818 176 2131 858 8082 7961 6756 7102 4949 7409 7265 4697 3592 2868 3364 1147 3103 7486 4025 689 7091 2436 7233 4513 3719 3298 745 7519 8119 7166 5836 6353 6778 6176 4211 2399 5117 2830 4423 7609 3399 3773 23 604 4305 1133 2260 7450 939 4514 4717 5523 3842 353 5050 7784 7495 6966 5560 4326 5210 3953 1479 2606 3334 6861 2114 1259 2501 2076 4891 5035 6213 5220 6707 224 6662 3718 6209 7782 1573 3856 6357 7989 2018 1382 4220 5866 7867 625 328 2756 6342 1549 4478 5266 4493 5184 1279 72 4379 456 7868 6266 6834 6638 1555 2312 3612 3694 6752 2982 6312 295 186 5907 7123 480 1346 2196 8063 2220 2556 1468 160 3073 7026 7507 2371 1059 2211 6681 4673 122 6779 7879 1806 5194 937 4844 5485 2651 7080 4876 4928 7869 2162 3174 7777 241 7200 5404 7151 2268 3814 3938 1275 2069 94 6234 227 6009 340 6931 3847 3538 4016 481 496 3266 2583 5385 917 8055 1415 2309 3028 7677 2466 5835 2383 5467 696 6657 3147 5534 3749 436 2828 6758 4294 3663 6683 4235 2793 238 3140 4481 5089 5118 2230 88 7752 3305 5039 5938 3145 6468 4430 2548 5784 6268 2557 944 1537 5024 5881 2566 5995 5459 1274 5967 6714 3074 7902 6645 5925 5165 6699 5268 7845 7366 4787 4913 5482 1377 884 4480 3027 2094 4818 5079 1030 5545 4700 6836 3599 368 2966 5074 1848 1936 987 2543 4193 1169 5207 5740 820 2947 5944 8132 1615 1130 1334 6391 7230 2338 4047 6664 2922 1117 407 4880 4602 5316 3121 5054 285 2981 1292 5883 5016 7275 4257 7709 1443 5306 7653 4030 5299 3720 7225 3496 3128 6396 3646 6258 978 3639 7052 4409 2187 7567 2820 7344 2416 1224 6924 3525 3643 2792 3914 6900 3209 1926 6642 973 2992 1246 102 2785 5375 314 2676 3131 687 5810 6589 4322 3022 417 1040 2800 7471 2847 2939 3041 6173 22 5804 7727 4416 3427 1103 1829 5260 5592 3221 7531 6154 5886 5179 325 7099 6202 600 246 5671 374 1446 527 6144 6424 3723 4440 2958 1242 4791 1200 5084 195 797 6660 4144 3905 7126 5028 2639 3896 4172 5819 1404 4093 7798 4971 1454 7257 5132 8006 103 3541 1236 6872 2324 1086 2742 2263 451 6737 3093 1286 6301 6952 1915 1386 5623 6804 5964 2493 4272 1699 4367 6536 7260 2623 1164 2668 3490 259 1903 4298 1260 248 3283 6866 8090 5754 2834 6499 4426 2684 4780 8073 6466 349 4484 7884 3693 2401 2894 4964 616 7147 2005 6978 7170 8083 6178 1888 1427 1128 3734 5257 1095 7248 3384 6406 7558 5658 432 1192 6598 2718 1312 5287 1011 6042 2390 2695 7390 4939 2765 3635 596 976 2471 7094 1104 4217 3830 2825 1822 3924 7199 5059 5721 7488 3020 3683 4586 8102 338 2480 4453 6777 1254 7038 5666 6092 5282 5831 3441 1923 5915 3807 4763 7760 1340 6214 6708 4723 4476 3982 7439 129 1105 8110 7872 623 6792 2325 7253 3676 998 4062 7046 6161 5 6684 4248 1748 5116 1984 3352 1749 2748 3557 6267 6027 7404 4131 2392 673 2303 6821 1636 7336 6544 3841 249 6625 7185 3795 3966 7580 5064 2919 3780 5175 6395 4309 984 2848 5738 1823 8064 929 4655 1176 5180 2472 6896 7428 5303 1387 6783 5008 2381 5190 5644 161 4341 838 7103 5507 1421 2258 5811 5746 5473 2689 5966 263 4914 6742 1959 4881 3045 3554 2760 4215 6784 2154 4792 85 4284 2231 2379 8019 5977 1396 8023 1057 6604 6576 1775 1464 1174 3606 4005 1284 6693 2661 6982 7120 6827 3756 2109 7790 3988 5153 6720 3257 7540 3641 5072 5736 7317 4789 468 4921 2432 2672 1229 2132 4140 402 421 5000 4873 2987 6105 2116 3127 1620 7575 7702 3735 7455 4106 4542 3382 1062 6420 4674 309 1916 4413 2066 191 1129 1301 3086 7749 4841 4757 1545 6464 7264 5852 281 2761 1409 1127 1186 5087 5204 3297 6908 3726 3296 2971 3070 731 5536 2779 4642 205 5639 6882 6382 3536 7871 4779 1481 4337 7303 7195 3906 3650 3240 3829 1508 3136 1043 4194 7764 6008 5797 4747 1970 2965 485 5559 4139 217 1721 3487 6270 6520 5637 1195 577 7479 7585 1925 3828 7936 7850 8085 997 7728 569 3436 7652 7075 2118 1035 5900 5224 3265 3817 2657 253 273 7070 1773 7627 4631 1595 5125 7696 476 4944 5398 158 5803 287 2652 3066 4688 5377 4356 6551 2183 7278 7672 1194 1544 6034 7908 236 7101 2601 4122 6465 7830 7644 7866 4463 1654 2572 6276 6649 7498 7969 1629 1730 6811 6115 5353 7812 106 4671 2713 6452 655 7772 5458 4046 6483 4517 1102 3062 3250 4755 6937 7693 3462 1790 7811 846 7000 3517 3238 2967 405 1606 6050 8053 7316 166 5978 1565 1107 4067 7217 2489 1843 1852 6992 3481 6675 6815 4754 4483 5168 1619 7511 3223 3370 4729 1611 5289 6294 4150 3827 6991 4474 1754 2149 7895 6356 7319 2359 3402 8092 1626 4982 7837 4805 6691 1449 4994 500 2885 7174 5145 6223 1502 8120 4526 7041 2157 461 7012 7722 1318 925 4706 4007 7307 789 2985 5052 6509 4422 1694 2304 6976 790 7545 1219 6975 3450 300 1891 4374 717 3626 192 1817 7576 7039 2451 6513 2082 6244 814 6366 6001 4715 2478 498 2081 222 3362 6646 4346 1527 1250 4375 4078 6967 286 740 953 1977 7639 1610 7011 211 8079 491 2685 3739 6909 1160 533 4350 4345 6685 6743 6502 115 633 5499 6963 3346 5177 2335 6766 1753 387 5086 6987 5196 7045 636 3332 2712 5890 724 2678 6171 6086 956 830 1530 4090 3015 6744 4941 1354 3742 5802 4538 5412 7522 1995 5082 3200 1666 2984 89 6435 3882 4165 3418 2527 1201 7138 5362 2479 7655 2034 1819 3187 6287 425 2941 8016 4467 5367 2898 38 7594 3245 7125 668 3013 4860 2055 229 6560 6233 1132 7841 2771 6822 188 3345 8072 4953 3877 5387 5873 6775 6647 1490 1393 6799 2709 3 4425 6141 5722 1764 5182 5962 7683 4012 582 4071 883 6403 8127 2476 1263 553 1717 2729 3910 4335 2644 7196 4738 1290 223 4148 5600 1175 7599 3411 4333 2358 5525 3550 2266 5697 1883 2983 2659 4270 912 5630 6526 6225 5624 5192 482 763 3762 6459 4549 2994 25 1735 7747 2964 4640 2464 7734 7634 7898 7586 7340 3183 7957 4831 765 5349 2409 525 5764 5342 2239 2907 7334 2554 568 5854 6635 4228 706 3142 3470 5768 1574 4704 1571 291 6067 1742 4916 5668 6063 1968 7718 6761 5090 6769 6831 2164 5663 5816 647 6149 6800 2980 5933 2660 2719 575 7292 2278 5130 845 3597 7281 364 1689 6878 1713 7444 5060 6261 3608 6315 4253 1258 2378 1440 7614 3777 2995 4002 4210 5142 1111 5959 4276 6538 2553 3007 3244 1311 7192 4598 1718 6273 8067 896 1928 494 6648 7429 6582 3873 2972 6533 6495 4677 3176 7135 2147 6239 3989 4651 5433 4490 609 6972 863 1522 6820 1663 1321 3423 5598 677 3567 6328 6492 1509 350 4853 1234 7294 216 7539 4957 4556 6414 3309 6843 2812 3904 8024 1640 2867 6630 15 5308 746 3137 1013 7758 2536 1839 7613 6359 4615 3960 2403 3243 1969 5135 7974 6981 3342 6183 2517 3286 1791 4616 6384 2859 2197 250 1592 131 4315 900 1368 792 1065 4707 5470 3444 3680 1304 857 5538 6651 1935 2900 2877 4577 2045 7663 3839 317 2026 657 7826 834 1151 6894 6605 7037 7579 8025 6767 3229 4347 4343 645 8032 477 293 119 6381 959 2899 6922 5599 6603 8126 13 4384 1310 77 773 19 265 2927 6200 5320 7904 7932 235 3125 7700 5201 1605 2530 5457 7616 4911 3949 6850 2375 7694 2013 3412 511 5893 4685 2138 136 6433 7293 5807 2112 5596 827 7414 3813 4156 6033 93 1532 843 2453 2457 5245 6022 877 1986 6736 5463 6593 7924 128 3972 4283 599 7327 1635 5270 3767 1759 4202 157 2903 6334 514 6971 3815 7650 2989 7066 4503 5359 4266 5800 6431 4888 4355 2988 7406 7371 4595 279 5253 7427 1276 3900 6697 7952 1788 3634 6833 2803 7854 7674 1514 751 7619 6668 6807 2938 233 6003 5606 4739 7024 3678 2021 321 4962 3601 2674 5114 3374 120 3186 3111 6278 5251 1723 6335 6212 1028 7565 3530 6044 1863 179 6235 6689 2633 3016 4262 6949 4055 1014 3413 3216 855 1442 3837 7569 247 3614 33 6888 5355 2608 3987 4365 2686 7173 3576 5264 1752 7899 806 6486 6612 4084 2795 1835 4142 180 2901 3671 7690 5561 7822 5837 632 5222 5274 7636 4784 894 290 1225 7464 4587 7216 4915 6734 6053 6347 3879 3264 3372 7048 702 2017 7443 7954 7461 3000 6997 3375 3860 2624 6291 5379 5148 1216 2512 6319 3152 3317 1071 7880 3613 5657 6557 2328 1271 4588 3386 4479 2653 1353 7776 7381 2192 7813 5555 6863 7098 3954 5672 3824 4867 2387 2845 8038 6255 6422 2533 6643 672 7562 4720 306 5197 6617 4291 650 1338 4732 4995 4902 4088 5794 3909 5781 4307 3627 720 1854 5214 7935 8087 6240 3042 5524 4955 6389 7692 2580 6090 3199 3791 7780 7946 8098 2421 2176 1150 3569 6694 7057 1524 3928 738 4749 4019 990 6701 2031 5712 1712 3746 1677 2287 399 2862 3051 2881 1801 385 5827 4520 4785 2368 59 7577 3929 709 1073 886 4532 3692 5954 1876 1372 4263 3672 5741 413 1869 2878 4371 1581 2354 5040 4756 4225 4227 4752 2937 5139 3081 4848 5589 6373 1874 7239 6363 1444 401 5088 8122 6087 3280 7282 1036 6179 3048 3308 2778 339 7349 4386 2936 2365 2621 7355 6263 666 256 7440 7049 6919 1518 4069 6430 1623 2819 7705 3212 6629 4489 6729 3930 6132 7857 360 3340 60 5715 5094 2305 7711 2056 1145 6127 7628 1867 7268 1650 3755 1087 3233 872 4036 6469 5075 6232 1971 561 5392 7208 1567 5200 5302 2477 3172 7315 2454 1618 544 7524 5939 1842 1578 2053 3488 780 5806 305 4713 1148 4256 6229 7411 2323 470 1755 4748 57 5045 4553 7119 6061 7100 5134 1722 1230 2111 7855 2708 5128 1359 2886 4560 2754 5085 4870 4289 5843 8028 2815 7244 6550 2776 2585 6755 4 821 2159 3426 7410 2957 4590 4561 2286 4456 5690 5106 4128 2611 6302 7836 3343 5799 7279 2339 6081 4049 2102 4892 4076 2694 5675 3302 7042 603 3472 7209 7193 5937 5645 4056 371 2730 7288 2057 12 7394 4487 3089 1188 1743 5479 1802 5895 5850 624 871 4293 518 699 5167 4555 5670 5817 1294 5092 7535 2915 6219 7805 875 5662 2243 1126 4087 5918 1535 2178 6032 1283 4837 952 4889 7358 8052 3565 4942 418 5083 130 7129 2649 5356 715 7206 6023 7588 7552 1933 1568 3446 7145 2349 8047 7564 1255 5579 621 7986 6591 1781 4753 7083 2245 416 1082 3497 173 6619 395 1830 2755 5506 6191 8008 5382 1562 5618 723 6153 3227 7324 2022 671 3522 5980 1198 1425 7919 7142 5228 902 6165 4761 2823 5871 4978 7765 4245 2264 3195 3281 4898 74 5770 6329 2494 2507 4539 3156 4432 2491 4448 2648 2405 2033 712 4940 2565 5947 3508 5469 4053 6814 7009 1098 2169 5988 6168 2389 5834 6131 5680 3761 2077 7458 6323 1622 6789 854 7638 1886 7017 4417 1614 7314 1846 5419 4204 7022 5311 901 2552 316 6037 3892 2088 1657 2216 879 3556 6958 651 7906 5372 34 5198 431 5144 7061 3217 6907 7888 6803 3887 164 472 2914 1414 3368 1314 5654 8106 5422 6705 4798 1376 8005 3104 3228 4630 2602 4196 2667 7423 3782 4833 2650 5707 6772 5007 4866 6375 4608 1538 8123 2854 3722 1505 6025 719 7008 4694 3571 4766 2515 7389 6673 6062 5511 21 2764 2308 2568 5068 6911 6874 2738 6925 825 966 2199 1096 356 1521 3029 5345 5699 1266 3482 1757 3234 4015 4583 704 2524 3478 3850 5193 3366 4793 3733 6892 258 358 4954 326 3394 2626 4534 7508 7828 4629 7651 5904 3880 7467 3551 809 6111 6257 7901 2540 6207 6135 463 2129 2190 8039 6956 3096 3078 4846 7710 6174 1307 4938 4502 5548 5494 8116 6344 4774 7916 3729 7900 3495 4457 2711 7480 6721 156 7773 2745 727 5679 6750 7004 1536 2101 4993 4899 4292 1146 5638 2855 4001 125 6085 7744 5163 2425 5317 816 7658 626 2218 837 3810 6628 4637 232 5812 6973 4772 383 7697 466 7002 4077 2306 5958 2038 5161 1010 4236 3380 7035 920 3101 251 5994 5649 2768 2307 1950 2126 292 7296 5149 3645 3659 5108 276 4842 899 6548 4300 2330 6259 2882 1004 5448 6177 7559 690 4981 6719 5825 1648 4501 1485 5546 6998 3057 7733 2889 7929 8045 3460 4000 848 398 3965 762 2733 736 7645 7973 3270 2060 6028 4523 4535 2805 2991 750 4166 1840 7937 6104 1178 3983 7771 5036 8033 4675 7602 2223 201 7148 3139 1868 2635 4442 40 7318 3307 970 6867 4568 1155 1750 3459 4455 7635 2887 4126 643 6936 606 1600 7737 2161 5792 4207 7489 5020 2186 4593 225 5815 5486 7065 6052 2280 6411 5051 2840 6107 478 7089 7457 2046 4190 1972 272 2240 2801 6709 5212 5608 3798 2798 3893 4887 1434 5984 2619 3106 1732 278 7953 3548 3872 1688 423 4696 3657 4620 4387 1352 1798 6094 6074 3716 1947 7701 382 4368 8099 1938 6558 4975 5124 484 7834 1956 7987 4138 5003 4562 4970 4226 644 1949 3179 3895 6068 7556 5586 1767 2762 282 5572 3899 8000 459 3247 2068 165 5887 565 7695 4485 1872 7393 5909 4299 4855 7863 3662 7212 2469 3688 4927 1371 6481 3704 8070 6616 2142 6599 1541 98 50 3653 228 629 5514 1583 4125 6139 4653 6285 3803 7304 4663 4254 2484 2352 92 5539 2468 7214 4464 8089 2233 5243 5378 7909 1740 1020 5332 5096 7133 3875 7649 3435 6763 6238 3341 5496 2831 3804 7980 8027 437 3514 6588 2291 3050 6703 4736 6696 396 4169 7894 4665 4743 1170 381 3528 2954 1940 7703 7399 3422 4666 5552 2418 1558 2537 1892 4618 4839 992 146 1596 2423 6819 3758 6364 3001 6690 7505 6436 4726 4510 2769 7122 1322 3367 4080 1375 7130 3052 782 1639 1461 5568 1763 6113 6626 3035 3849 5906 3925 7930 475 6686 2506 774 3295 97 2460 2028 2826 4354 6117 2664 4394 4682 5259 1924 3331 1185 5215 2299 242 1656 145 4278 3456 4708 1362 6842 4803 3032 3222 6859 2437 2592 4129 373 5725 1744 4933 1045 2448 2949 5676 793 140 2215 4267 4885 2424 7416 4110 6013 2012 4206 7931 5902 7563 5235 4059 1403 3714 3420 181 2431 1070 3377 1408 3162 880 1690 803 7132 1772 5976 3314 7755 7387 2541 2194 4118 448 4044 3454 6611 1962 7646 7361 4790 5246 7617 5483 3623 3360 3743 1054 2597 369 422 3685 3572 6886 5221 6933 3207 1476 2340 6476 6376 4406 1063 7933 7988 2121 788 2641 7312 6313 7548 384 7527 5189 1177 2357 6030 301 2827 7551 3339 5692 3967 7301 935 5025 4735 4018 1480 7064 1710 5310 571 7667 8041 3901 6573 203 6325 1559 288 5076 2956 7143 5103 30 6455 3871 4421 7940 1451 5333 1217 7420 104 3087 5101 1871 2221 2876 2269 1976 2716 3959 8036 3570 7447 7779 653 5147 3054 5580 5501 1889 3079 1594 6718 3213 7595 6559 4151 578 7666 5841 6106 7013 7221 1849 3173 3105 2486 1511 5138 344 6798 4243 4804 4363 2902 6893 6887 1222 6350 3434 8001 6463 7890 1746 6045 5682 7347 4482 7084 8030 1931 4486 5713 6479 5455 1108 7823 183 1232 5626 6671 542 2680 299 2041 1337 3964 1143 3956 5369 5031 4599 3461 2654 3573 945 1510 2632 2146 3962 2337 7622 6222 1472 2413 5183 3149 5650 2978 7305 7637 6845 3451 6838 4961 6125 3401 537 6454 7144 2561 7883 5826 312 7544 6840 7459 950 1076 5357 73 7251 2740 4325 4115 1582 5872 2415 3019 1828 7020 1227 6108 4466 5098 3581 5878 7116 1517 5557 6515 2473 6058 209 3166 7470 6946 4912 7168 5322 7896 748 7561 6011 4119 4973 5329 2289 1528 7107 1957 3135 7269 5735 5602 7263 2293 4102 2350 3242 1106 155 4471 5605 8050 362 3687 5917 237 5374 1953 1115 6150 5787 7492 1932 3738 151 1813 5703 7456 5851 3463 6121 551 1922 662 7202 6994 2579 1504 2067 2171 269 2534 3077 5597 5695 7201 2767 1426 6036 2974 1373 2959 7706 3021 2433 330 6110 5241 3516 3741 2201 1994 1083 3320 3447 6236 1319 6546 4499 5716 2091 3047 4342 3927 471 5223 2314 262 3590 4187 5551 7096 4872 4571 1946 20 1864 7432 2090 1291 3203 726 4851 1088 63 6091 7601 1553 1821 2909 6607 1477 7176 3973 6418 3764 6447 6016 2127 1877 3633 7999 737 2707 4572 3539 1090 446 4428 5704 664 6308 1729 5037 7290 995 1328 6029 4969 6856 8133 4923 1183 6330 3193 3473 1686 7654 6166 5026 5979 7881 5150 1012 7978 231 3160 3058 6457 639 2404 539 5300 6797 2759 3330 1765 6738 6915 7280 7624 649 7386 619 11 4357 2968 5151 6054 1634 2420 6898 3038 7376 6768 6524 7478 5454 2577 7299 4654 4279 4452 3751 5607 2986 7465 2054 6883 4877 2872 4108 4198 3684 5043 831 4362 8107 3067 6613 2257 2070 2884 1390 3625 3577 2 7460 2391 5952 7704 2115 5478 5846 4709 1858 204 3936 7329 968 4120 4130 4314 7631 2806 3485 318 7391 1297 6084 5789 598 4930 1885 4237 3985 5992 813 7114 4632 1167 474 5745 7333 7785 4295 1631 3503 1906 1784 4823 3285 5849 3359 882 7223 1638 7514 5780 4582 5453 1370 2816 6147 4825 1406 444 5894 6623 6248 7927 4658 435 1288 7364 2582 82 3276 2298 7566 8057 6197 3396 2871 6594 4541 7302 6194 2253 6425 8129 787 7547 6858 6531 5793 3919 7840 536 3818 7003 5729 586 1585 1355 2217 4945 42 2732 6817 4286 2734 2634 4184 7550 7642 7597 6155 6722 6429 922 2589 1249 6383 2945 2386 3182 7818 1996 147 5335 4449 1613 2799 1032 2600 779 1204 4065 3848 3958 449 3230 1364 4275 6554 7350 6579 3337 4826 2837 1988 3857 4369 100 4223 6810 2725 313 3351 7071 2004 8137 4258 6869 2786 1684 5490 2106 4164 3533 4037 7093 5456 5923 1182 2890 5576 8086 6849 1006 3259 7921 2288 90 2406 4731 6336 7342 6592 772 7815 2908 3686 5930 4937 1273 7019 567 1498 5230 7 1203 4762 354 5269 703 4068 1241 601 3198 8029 5283 1692 3664 5717 7187 2797 3748 5786 1315 5683 4306 7778 6064 3816 1855 6193 5515 4153 5127 7675 4495 559 2210 6916 2023 4419 2843 3785 1711 7337 1647 132 6757 7592 7997 4104 7630 7922 7721 4865 1075 1038 1316 6725 6764 5160 5304 1056 974 6122 6419 2474 2117 7793 915 280 1708 283 365 2246 1904 3090 4999 4521 4447 965 4769 7398 5842 2631 714 7657 35 1627 6145 380 7792 1395 6427 3046 1388 3725 3389 6665 798 7970 118 6751 1811 562 5655 3836 903 4100 7186 6354 5360 408 501 1413 6741 2050 6547 2860 5365 5863 8058 1825 5824 4594 3876 5615 8130 1560 6620 5111 6300 7981 4771 4137 5231 5505 4092 3952 4863 2475 5718 3709 1513 4444 4212 2447 5884 1416 335 3063 7541 587 6010 6441 6539 7647 2087 791 1897 3175 5629 6473 3787 5653 7875 734 3765 7194 2663 3835 4584 2119 1162 6205 1438 5739 7149 2398 1714 1992 7796 6245 5435 1041 81 3990 3304 7322 6512 1561 4827 6514 5973 958 6252 1347 6124 6634 5229 6134 5969 1715 7515 4060 7963 692 1601 2976 1516 7341 759 1556 3506 4096 5687 6537 7861 1313 2241 3652 1448 2560 6947 424 4764 2274 6160 822 7807 2347 7481 4475 8096 5030 4909 2807 4636 5640 3881 1768 2775 5381 4420 7757 1810 6677 2459 2043 1674 2006 3537 4591 2809 3840 5341 135 3698 3256 219 148 4114 5277 3279 6450 4659 928 7226 4038 7708 3637 4906 2931 2205 2318 735 5481 5919 4074 2586 3133 5427 801 6242 1789 5856 3763 2620 1462 6754 4296 2875 2656 2397 951 2538 5249 3059 2065 5526 4519 6390 1017 594 4135 7799 2563 7047 5619 2593 1029 6572 7789 3661 2516 1358 2032 5104 1207 2071 4943 83 1724 7286 7848 4111 1270 2610 7525 4657 189 1016 2810 2741 5199 6097 627 1015 7949 3237 4500 7555 2814 685 429 4123 1079 4626 2706 8075 1832 2675 1512 6477 5418 7877 4461 7382 4045 4436 2219 3159 4931 3679 4613 1137 1652 2904 3387 1048 6321 4199 2677 6470 3405 5129 2896 4250 1591 6897 7538 1110 6786 5105 718 2213 1661 6853 6639 1589 7928 2570 6487 6844 5272 2929 7536 3644 3778 989 4431 7025 7357 2500 3191 6600 377 5186 5091 5991 4261 6695 8021 7892 6369 3717 4746 852 7088 1109 808 392 1452 2110 7606 584 2594 2200 4388 7947 6189 1327 6955 6806 4469 4681 7945 7804 4811 5009 1609 991 1131 680 5286 1704 6341 2051 1424 1814 3453 7162 5867 4255 7321 1665 3011 6940 4987 5157 3403 7510 2255 6661 6079 729 3419 1053 7753 39 3271 7809 3357 5681 1374 7256 893 6918 1844 3388 6275 208 488 4882 1478 6713 6164 6 2513 7914 7402 2086 1907 1702 3113 6311 3371 3205 460 4992 9 5577 412 7907 5202 3088 8017 7252 2564 400 4174 4437 2061 1697 1055 3542 5430 3084 7846 4327 7378 4879 4338 5756 6047 2214 5370 2311 4646 5424 1399 6902 869 1369 6421 2804 3832 5121 1978 3168 2203 329 7917 3004 3126 7157 3180 6306 3117 3468 6943 6928 4660 4200 3997 3115 2042 1653 3369 6578 2342 5727 2951 5971 245 7014 6345 5046 6711 5723 2481 8043 3442 5914 4330 6407 2247 5019 5158 6552 4424 4397 8013 932 2510 5475 5340 3529 1807 185 8062 3585 3024 1630 6456 5882 927 870 6496 6774 7743 1269 1834 3385 6413 851 1682 5056 705 7167 7502 4445 5063 6812 833 1707 3543 3995 6157 7043 4730 4054 3178 1136 6020 7408 3146 1974 3999 5858 2402 6773 4247 905 675 1420 323 6566 6046 3940 4336 5033 700 2627 2883 3082 1989 4948 4042 6700 5408 7500 1238 1039 4435 7097 5609 760 1008 3449 2528 2503 7615 3610 5633 7763 4724 3475 6133 4359 3609 1646 3039 5006 3118 193 4400 6324 1021 2396 6137 3110 8080 8135 1910 5460 5368 3917 683 5029 6771 6682 2588 6041 5772 3638 3942 981 3647 123 1870 5275 4066 1804 904 6654 5281 4932 4799 564 634 3915 7991 2229 4649 6901 2488 4221 1342 6217 3912 4814 4742 1728 3630 3715 5252 2297 2275 5434 4834 7918 5250 1944 6727 6143 2757 5956 4875 5935 406 5889 1134 483 602 6265 3313 5237 4917 5057 1554 3326 3560 6723 3060 8113 2446 3112 3768 7783 7261 1751 1987 319 1700 4460 6602 971 1067 4810 4319 1378 5227 5131 4956 2508 1064 6043 1678 5823 6181 4488 6066 310 5174 1795 5808 2265 7554 2605 61 555 7629 1576 4910 1683 7291 4565 6226 1257 6828 1303 5847 2009 2703 6846 3591 1218 4238 5822 1529 3945 4528 2441 5325 2030 4559 4098 5073 4009 6386 1797 3547 1703 2499 6852 4647 1500 7405 1384 2395 2784 2195 5759 749 6355 1488 3891 3005 2388 4540 4473 2224 2522 1302 6326 5425 1231 4908 6829 5796 6439 7860 342 3323 3642 2924 3378 4868 1685 4162 4744 7994 6218 1379 6974 811 2691 65 3202 1050 252 5693 1706 812 3143 367 2208 663 4667 7671 4101 2996 6112 6158 3316 1853 4052 7345 796 3864 2704 6930 6443 4017 3361 1211 7392 3500 5970 4124 3846 6702 4412 4656 517 7323 888 7053 711 4543 3363 4668 75 5366 2362 1407 4321 1140 4531 1343 6277 7325 794 7820 6051 4901 3833 6880 6077 7073 4806 7058 7388 1119 359 1320 3870 2360 4376 7959 652 1599 5276 3681 3255 4711 2374 1350 1401 5484 3632 284 3978 1698 6522 2979 4492 2960 1060 1668 3579 5531 4496 7750 3951 628 2002 2384 6057 5452 678 1348 1210 4575 2084 6453 5159 2744 4304 6704 3251 2923 6378 510 6762 5402 5669 7134 6146 1861 3772 397 3531 47 2891 3430 184 826 4033 5612 547 2079 6717 686 3479 214 2917 1061 2462 3695 2715 1691 3061 3458 1332 6879 6523 6428 4089 7018 2317 7738 3409 6307 3116 5284 112 3578 3410 1152 6005 6787 473 2093 6471 5921 2973 1325 6730 7521 1964 7110 7716 5591 5961 7516 581 850 6251 2682 7442 2906 6416 2408 5694 688 5573 908 6399 4548 1467 4269 1880 7218 8040 5942 3277 5998 2175 2204 4287 4597 438 1483 7175 3826 6624 5122 3898 3055 492 2440 916 455 4604 4116 914 1809 2628 1927 2724 8091 5928 7027 391 3660 5405 2355 3009 4664 452 572 2687 5965 3843 8014 7413 3018 4634 6126 6813 3934 7827 2692 6412 1190 6444 3356 7882 6747 4529 3918 3263 5280 758 270 6475 6289 3438 7530 1607 3675 4991 1453 2155 1037 4378 2256 6999 693 1240 5348 1887 2177 4392 2495 3157 5169 7950 1125 4801 1645 910 2751 1405 5728 7851 5940 4617 187 3012 648 6525 24 355 3802 5862 3321 3861 56 5661 162 4795 8037 4185 3037 4947 1954 6247 7182 864 6500 159 8104 5566 1624 3181 2782 1361 6075 3443 6480 7418 2612 4639 3476 220 4331 6595 5575 7276 3712 7076 2095 1985 4351 7105 1208 807 5604 3600 7445 2607 6746 6862 5468 7023 6039 6083 2852 6716 479 361 7942 955 3099 6816 5913 7596 6368 1850 2444 6130 7131 3507 2207 1493 5861 3631 7600 5014 1094 6310 7870 5509 7172 3667 5005 1499 2546 7983 3916 6505 728 5719 6609 6534 5766 7415 7155 5390 5061 4061 5750 6410 7368 1163 8105 4401 239 5141 3164 7338 1741 5544 6622 7188 2944 3526 2372 5931 2458 7412 6659 7732 3957 5957 4852 4297 4332 5256 409 3788 6076 7487 5099 3731 658 3894 3192 3194 4733 5777 5062 942 2613 5855 6279 841 4670 3318 6904 5801 6088 4443 127 4022 4339 2236 4010 7520 3808 1779 6241 6460 6655 4418 1771 2428 4323 8054 1818 5516 3654 768 2092 7277 605 5238 3246 6575 5710 7625 7219 4854 4537 3674 4188 4545 975 1519 2336 7179 2911 363 2870 4566 620 3796 1473 1866 1003 6501 6788 4765 4404 3673 521 6780 3030 2113 786 4683 6096 138 212 7473 4990 3993 4645 3555 489 6048 960 1705 7295 7874 334 58 3805 5897 4786 4816 4201 5386 4155 5642 4689 3319 5376 1919 4624 1245 4029 1389 7910 414 4840 3083 6089 4167 5951 6571 7911 3151 3948 5689 2327 2780 7832 7451 4862 5391 6449 6012 4619 1248 4348 1566 3201 1187 895 352 3350 5407 427 1993 2609 2946 3690 7532 2829 7852 3226 4550 3518 5955 2099 866 5285 3984 5848 7720 4157 5081 6237 1506 2880 7006 5620 6969 1199 4161 4458 982 6019 1287 4569 7598 6938 6451 3923 7081 4438 1826 7436 2296 2130 585 2407 2705 1118 3568 4182 4107 221 2150 5926 3908 1402 7234 1025 1005 2646 878 5315 7215 887 3656 7668 4334 4203 6365 7802 1785 5010 4919 388 890 2850 2598 1436 5017 3220 1681 5263 433 7593 3071 4644 5004 7513 6953 1805 5480 3471 7678 4554 2681 1898 3523 836 7431 5760 3598 4006 1902 7591 26 6385 7965 297 62 3491 1580 4924 5465 107 3588 274 5588 3797 4558 7955 6614 5438 1649 240 1237 6303 3043 4064 3902 2861 458 2910 6921 3852 1859 4934 2970 5318 5594 2140 5347 7862 8068 6951 5540 3844 4259 7643 5109 4050 2276 579 3189 509 4776 4213 7016 5154 7360 679 2281 1252 4621 3241 1760 4983 1546 3562 7641 5313 5038 4381 5565 2934 7687 7574 2020 5737 3208 5015 116 6269 5974 7707 4869 3056 2445 2963 4233 4835 832 5327 2576 1890 5126 3628 2590 348 376 303 6095 7993 4403 36 7285 5790 1680 7171 7944 1865 6585 6018 1981 5948 7365 1934 415 7090 6080 1471 6172 934 2888 2822 784 4972 1422 1921 2001 6692 1435 495 3775 1878 4635 6186 7797 7926 716 4896 4893 66 8093 7633 747 2103 5652 5975 4459 69 1429 7669 4699 7241 6404 7680 5569 7228 4998 7920 761 532 861 4160 5859 8044 2669 5324 5870 3031 936 3094 6832 7714 5832 868 2283 7506 1418 1671 5176 1142 7501 2567 6283 3968 1122 4405 1221 1253 5344 6653 5610 7079 3150 308 6749 5137 5567 4864 5763 357 6688 6393 7343 7664 7962 2735 2662 1282 2615 5865 502 7607 3325 1123 6148 4353 7587 1475 1550 7791 3903 2720 7156 4170 126 5500 3992 2770 5945 2781 5990 1423 1153 3868 1220 613 2492 4429 2189 4273 7426 1411 519 3534 7283 4702 818 4745 7005 322 1658 6781 6984 6570 7274 202 785 2637 78 5635 4672 2016 3416 4075 2835 7751 4472 7756 3790 5755 4281 550 3274 8103 2766 6472 5123 924 84 5487 4737 487 1920 1824 1780 277 1139 5321 3689 3068 6136 2234 6490 1979 4494 2074 154 6663 6793 8061 167 2697 6243 7483 3558 6790 3154 302 7180 6281 6462 3464 7518 804 1812 6556 4857 6782 3406 5497 2443 642 5776 1084 2063 347 1433 2833 7326 2153 2143 1300 7015 4603 7886 5829 3784 2584 5292 1808 6959 4768 867 7189 4847 4197 3583 548 6296 6314 5314 4171 1329 2209 4676 2714 1731 3889 4518 426 6316 1179 1552 5876 2591 4251 1309 7339 7800 41 7715 7838 2295 5875 6206 1114 3395 2261 3036 6192 45 3980 6871 7849 630 8046 7313 1716 3649 622 2133 4146 6377 2439 2120 2085 2037 7885 8114 5558 7843 1277 7164 307 3498 2789 5053 5291 4628 1577 2049 268 4280 1223 1196 7990 1899 393 4168 570 3771 1905 3602 6199 8078 6072 983 3721 1901 534 1637 7331 198 721 6372 2642 1879 5651 254 178 6837 4935 6337 3897 6069 7448 3666 3158 4026 2332 3806 3996 1667 3324 5077 3123 3888 7379 2072 1670 6503 5628 7698 6910 3564 3204 7159 540 6873 6208 1272 5922 5603 5396 3563 1299 5521 4506 2163 4043 1458 3393 1437 1884 3404 7207 4324 4097 2250 2321 3177 6712 6530 610 6004 3702 3153 7374 5363 6983 7059 969 1026 4679 3711 3415 1072 5326 8031 5240 5899 6274 6026 493 1676 3114 940 2125 5769 5415 6939 1089 5093 8048 781 196 6841 5616 7717 2774 3593 6929 6211 2483 6128 1909 3561 4878 725 7905 3745 844 5044 1 7328 2363 3944 7222 3397 3290 4186 5371 5901 2498 4871 1324 6640 7054 8015 7139 819 3566 5562 5078 6932 6215 3935 4218 4976 3706 5343 4013 4821 5012 1341 4782 454 5294 588 2262 2975 294 7754 3779 4035 6801 7086 5406 7729 6387 6478 5724 4979 667 2168 612 8010 5814 7808 6970 526 7203 341 5152 949 1531 5530 79 5513 757 3994 7903 6482 394 7367 7889 1367 1439 2015 4883 4802 4824 885 5742 5749 1761 6903 3596 1495 6823 6035 5541 767 5474 7610 3100 4271 6388 2170 3338 5916 3703 2137 3408 5788 5647 4622 6118 1941 6280 7964 4950 5071 4527 6808 824 7865 6167 5162 7243 5595 7584 6099 1069 2787 6415 5373 5504 3049 5047 8108 7856 2400 111 1363 7626 4897 6015 2455 3306 2449 589 6926 3109 6980 6851 6508 823 271 7395 4610 5331 1540 3885 744 5431 3148 522 4959 2539 1349 51 4208 3432 2301 4968 4788 4759 7761 7612 5762 1097 1669 5146 3587 1881 1851 168 4360 5065 1838 3611 6221 3014 3064 1633 707 5667 2895 6374 7235 5767 1463 3845 4361 3552 4219 7430 1165 691 27 1758 7141 5247 2752 3484 6082 244 1539 4396 4627 1450 2427 6140 3467 6996 3511 453 2569 3937 6541 5857 2699 2932 6504 8042 3549 80 5828 2285 1793 1239 7227 6070 7433 2722 3699 5518 6881 6360 7824 907 2438 6169 1662 5743 29 7982 615 7472 6392 7272 2048 5932 175 4178 6715 5394 6488 6927 6394 5143 7435 2136 4504 1157 4329 5206 2727 5517 4926 7346 5528 7485 4205 5950 638 1917 3289 6785 7462 5380 961 5631 1074 3163 3586 3218 7258 2348 1233 7249 1044 6941 2962 7383 593 7878 520 7977 1161 7309 2394 3171 5711 6706 7676 6498 7509 327 4967 2179 6494 4741 6361 206 4021 847 3913 8121 6120 7835 4222 6517 1168 5522 3618 4719 2616 4214 2581 4703 2851 3138 3794 6400 3407 3235 6264 1616 439 5908 676 4734 5351 930 4470 503 4003 1625 141 114 6995 6985 3417 4048 2836 121 7589 775 3907 1120 7767 2625 1267 1184 1397 5943 641 1206 4274 18 3466 5213 7537 1945 538 4574 6760 6180 4366 3863 4014 8094 2547 2658 3504 3347 5874 2853 7150 7736 5258 2385 2618 96 4242 2089 2039 7746 4551 5337 7670 5634 3658 2412 4301 7724 7726 3926 4398 441 1586 1138 2166 5242 1066 1709 4349 817 4832 2273 6201 874 2599 1563 1774 5354 1027 8124 3970 4974 4383 3392 597 124 8111 1251 5441 5621 7242 2346 1171 1963 4721 4758 4966 440 576 1121 5271 1007 5779 7493 4181 5795 2442 7662 1570 5312 428 6162 2940 2122 1642 1296 5818 6549 1766 8 3981 6467 6606 6262 557 6271 5203 7106 6564 4985 4567 3665 6885 6220 8100 5986 7821 2926 194 5726 3284 733 7311 6298 7979 5233 1523 5295 2746 2772 5399 6123 4109 2920 4760 5032 2272 1501 5782 4099 656 1308 4183 386 5844 543 4192 7618 2096 6491 3920 3144 7523 4112 6748 3433 7768 3697 2643 315 3026 2011 7497 3336 1948 3224 6864 1470 1833 2467 5173 891 1428 2035 3769 5395 1990 6156 2673 2242 5674 6007 3239 4113 7104 3559 3188 3002 5646 3300 207 1474 2573 4095 8118 4177 5191 7266 6687 6333 778 3197 4611 8109 210 5172 3185 298 1719 5388 1787 4152 1588 7740 1261 7112 799 3801 3328 2487 4041 8060 7996 6510 6216 6586 1226 5542 5205 2040 8115 3584 2232 5185 1189 4277 3521 1643 5508 4163 954 7051 170 5492 2206 4240 3080 5533 113 403 1737 1837 3267 1400 694 1503 5946 3065 6100 3932 4372 2098 5178 2271 7169 7262 2025 7925 8095 55 6286 5983 1628 7377 2212 4195 3215 6728 4829 5022 3553 3376 2523 7590 3869 5840 2808 5290 4446 450 6968 5384 6796 4856 4310 5611 3862 1857 697 4433 4175 1149 7689 1820 5002 6615 1770 7806 4643 7739 6203 4612 8077 5188 4740 1873 6253 6338 4918 4984 5771 3545 2930 7270 2497 5112 898 1679 6934 554 1080 6461 469 4373 370 3955 6254 3069 6776 4796 304 6371 5785 2333 4505 7247 892 197 805 3134 7396 6627 7077 5905 1799 7463 1937 5428 8007 7407 190 6401 5761 434 255 6195 2277 2790 7400 4986 918 1212 4900 2144 46 4638 560 1031 2545 3776 2542 506 1738 4797 1862 1747 68 3831 2310 5136 4132 6847 5554 1091 7067 346 6954 665 6518 7348 996 3092 5960 7190 3268 3789 4282 3437 3519 7036 3669 752 3950 5706 7351 4965 771 4234 6507 6256 5279 3358 2645 375 5613 7362 5677 5765 8035 1675 7476 7424 5833 7603 3429 1696 6791 8081 3759 2578 6580 4117 1432 3713 5941 5529 7549 5892 2666 4625 3097 5048 2603 5898 6950 2167 8076 3040 2562 4576 2470 7240 7992 6656 7297 4922 3219 497 4633 6024 5097 1209 1113 5953 6540 4011 6434 3520 6574 7528 3603 2249 3510 7373 7177 8049 7971 4063 972 5477 7082 7491 6977 919 7370 3732 4725 2326 6906 4180 2737 5775 4530 546 7499 2073 6397 2791 1135 7787 4133 4925 2846 1845 4705 1366 2549 4684 1911 7158 5095 4678 4451 6899 7829 5049 174 1856 6735 556 7152 6056 4809 6912 2108 7441 2341 5510 1991 7853 1952 5305 6432 7686 4886 4352 1047 1002 6351 2874 7864 6339 1306 2410 99 5397 5401 3232 6014 2282 2701 7210 6986 7178 566 2463 7557 17 5267 7300 267 2292 2294 4777 2351 5535 6152 6561 5058 4083 2202 5641 3072 1085 5491 1154 3747 7934 3095 6184 1173 3249 64 3225 2832 933 800 2461 1081 4040 5248 6101 4669 4512 144 49 1339 516 7876 3489 95 1392 6731 1579 6905 3365 1672 6545 3911 926 152 5437 5070 5447 6055 4127 2237 4522 1494 3737 2777 7581 7995 3820 1326 6528 7847 7238 6770 6021 7967 7137 6676 3008 4813 4477 3141 3494 2010 8002 108 4393 3017 3770 921 7385 4693 7069 2544 6848 753 5100 7573 1543 464 4103 6485 1330 1023 5346 3710 4680 4498 1587 3701 5731 2047 243 1983 1617 4951 873 2665 4903 5352 2097 1745 3272 5011 1497 6802 16 390 5549 260 3299 1725 3636 7632 4121 2869 528 8074 5400 802 5426 3428 486 5409 2248 7113 2863 7810 7095 2817 2783 2414 4905 1331 530 5648 1457 5067 7984 5705 695 1762 1381 5543 7788 2617 2373 6637 5758 5733 2434 7356 7621 963 1918 4552 7259 2075 4980 8003 139 2270 7939 6370 6231 6093 7375 7287 2364 3629 2794 5461 4389 4173 199 1193 7359 1967 266 5156 3098 2228 6163 2952 7124 3823 1958 4454 419 5414 5488 741 2003 3728 2160 7085 3448 7893 3196 2950 1365 3492 2456 7504 7380 8134 2647 6944 3513 2284 6114 5622 5069 7583 7353 5520 5364 6535 7205 2948 5262 5839 2551 3867 1548 4570 4712 5208 5323 3655 6142 5339 4783 1101 4370 5845 7181 6989 6230 7438 4229 5080 1602 6078 5503 5686 2559 4320 7951 2688 5853 948 8084 5556 1659 1289 5838 1466 4858 7640 2329 5783 3023 7063 4034 7529 2504 7976 1803 6182 2743 4268 3349 7437 215 3190 859 7887 2630 7108 3604 5625 1733 4158 1172 7819 4507 6017 7623 5171 4815 109 1608 4031 2683 5659 6577 2535 6445 8066 7648 1664 2193 6945 1973 1213 5996 7213 5013 6305 4557 6610 6185 3941 6652 6865 6990 1459 2749 8097 7912 1256 1285 2382 2813 4936 6224 5466 4511 1280 5236 4794 1391 3822 6190 2758 7769 7956 1551 828 7620 7742 5778 117 1603 4232 6698 2550 4907 4414 6923 1687 7533 1557 7482 6920 829 5120 504 6835 5688 5209 6596 6666 7570 1792 2525 8112 4609 5987 6272 5563 7352 6590 3546 5578 1460 1202 6409 6438 6346 1345 2731 6458 5924 2739 941 1001 6102 311 3252 4650 7572 4311 5113 5684 2750 2912 3648 6059 1298 2376 5664 4845 1265 4241 3275 631 8026 3574 6196 5293 3390 6484 611 5685 2521 3425 53 6964 1228 6398 6260 4690 5968 3783 5449 3708 169 3757 4380 5334 5410 3214 7725 4027 512 505 2990 3236 3886 5805 2723 1998 6292 3373 698 264 3883 1980 5423 6440 1796 6332 4508 6824 3312 2450 4020 5476 6618 3483 7255 1590 105 1914 5502 378 7731 1068 4230 999 5963 595 337 3607 755 3457 4601 4318 5234 2921 3294 257 3273 6038 1572 4313 1534 964 4224 3269 3440 2080 5553 3184 5288 4581 7781 3509 7496 3124 4778 4385 6327 2571 2105 1951 4820 6367 7232 2059 7748 5709 261 710 2496 4079 7873 2173 5244 1394 5809 730 143 5700 5383 1419 5753 4086 2638 3979 6318 4260 1776 3750 4716 5614 1093 14 4216 2655 7774 2259 5403 962 3819 5903 6442 7140 3621 909 8069 5472 523 3261 2235 5278 2007 6745 3211 4727 6632 911 8136 7723 2145 86 985 3754 7688 2316 3947 3730 7759 5027 8056 2141 682 6049 4547 2100 2558 389 4988 5752 499 2024 2802 7250 5297 5660 5298 2104 2393 7419 2811 5993 2174 515 1701 4008 4402 2191 1042 2511 2302 4265 2622 4775 545 142 6724 372 563 1900 7699 7273 4808 5627 3301'
data = input.strip().split()
n = int(data[0])
pi = list(map(int, data[1:]))

lis = find_lis(pi)
lds = find_lds(pi)

print(' '.join(map(str, lis)))
print(' '.join(map(str, lds)))
```

```python
# Genome Assembly as Shortest Superstring
# Given: At most 50 DNA strings of approximately equal length, not exceeding 1 kbp, 
# in FASTA format (which represent reads deriving from the same strand of a single linear chromosome).
# The dataset is guaranteed to satisfy the following condition: there exists a unique way to reconstruct 
# the entire chromosome from these reads by gluing together pairs of reads that overlap 
# by more than half their length.
#
# Return: A shortest superstring containing all the given strings (thus corresponding to a 
# reconstructed chromosome).

def find_overlaps(s1, s2):
    """Finds maximum overlap between suffix s1 and prefix s2."""
    max_overlap = 0
    for i in range(1, min(len(s1), len(s2)) + 1):
        if s1[-i:] == s2[:i]:
            max_overlaps = i
    return max_overlaps

# alternative efficient recursive greedy
def assemble_reads(reads):
    """Greedily assemble reads into shortest superstring."""
    if not reads:
        return ""
        
    # start with the first read
    superstring = reads.pop(0)

    while reads:
        best_overlap = 0
        best_read = None
        best_index = -1

        for i, read in enumerate(reads):
            # check the overlap of superstring suffix with read prefix
            overlap = find_overlap(superstring, read)
            if overlap > best_overlap:
                best_overlap = overlap
                best_read = read
                best_index = i
                # also check if read can extend from the other side
                overlap_rev = find_overlap(read, superstring)
                if overlap_rev > best_overlap:
                    best_overlap = overlap_rev
                    best_read = read
                    best_index = i
        if best_read:
            # merge with best overlap
            superstring += best_read[best_overlap:]
            reads.pop(best_index)
        else:
            # fallback
            superstring += best_read or reads.pop(0)
            
    return superstring
                    
    




```
